# Combining all the tables

```python
%%sql
create table full_data as
SELECT f.product_id, f.listing_price, f.sale_price, f.discount, f.revenue,
       b.brand, t.last_visited, r.rating, r.reviews
FROM finance AS f
INNER JOIN brands AS b ON f.product_id = b.product_id
INNER JOIN traffic AS t ON f.product_id = t.product_id
INNER JOIN reviews AS r ON f.product_id = r.product_id
```

```python

%%sql
select * from full_data
```

```python
#finding total revenue
%%sql
select brand, round(sum(revenue),2) as total_revenue
from full_data
group by brand
order by total_revenue desc
LIMIT 2;
```

```python
%%sql
select *
from full_data
where listing_price is null;
```

# Categorizing the price range as budget, expensive, elite

```python
%%sql 
alter table full_data
add column price_range text;
```

```python
%%sql
update full_data
set price_range = case
    when listing_price <= 50 then 'budget'
    when listing_price > 50 and listing_price <= 150 then 'expensive'
    else 'elite'
end;
```

```python
%%sql
select * from full_data where price_range = "elite" limit 20;
```
